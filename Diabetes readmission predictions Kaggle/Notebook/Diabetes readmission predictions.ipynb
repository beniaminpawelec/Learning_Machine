```python
import pandas as pd
import numpy as np
np.random.seed(2019)

from ml_metrics import rmsle
from sklearn.dummy import DummyRegressor
```

# Imported Data

```python
train = pd.read_hdf('../data/diabetic_train.h5')
train.info()
```

# Basic model 

```python
feats = ['encounter_id']
X = train[feats].values
y = train['readmitted'].values

model = DummyRegressor(strategy='mean')
model.fit(X, y)
y_pred = model.predict(X)

rmsle(y, y_pred)
```

# Submit

```python
test = pd.read_hdf('../data/diabetic_test.h5')
X = test[feats].values

y_pred = model.predict(X)
```

# Submission

```python
test['readmitted'] = y_pred
test[ ['id', 'readmitted'] ].to_csv('../output/xgb_clf.csv', index=False)
```
